```python
import numpy as np
np.random.seed(5)
```

```python
array_1 = np.array([5,7,6,8,1,3,4], dtype=np.int32)
array_2 = np.random.randint(10,30,len(array_1))

print (array_1, array_2)

print (array_1.shape, array_2.shape)
print (type(array_1), array_1.dtype)
```

```python
print (array_1 + array_2)
```

multiplicación elemento elemento

```python
print (array_1 * array_2)
```

multiplicación punto a punto

```python
print (np.dot(array_1, array_2))
print (np.dot(array_2, array_1))
```

##procesamiento de imagenes


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = Image.open("/content/lena.jpg")
plt.imshow(image)
```

```python
print (type(image))
```

convertir ese objeto a un arreglo

```python
image = np.array(image)
print(type(image))
```

```python
plt.imshow(image)
print (image.shape)
```

colores minimo cero o máximo 255

```python
image = Image.open("/content/lena.jpg")
image = np.array(image)
#image[0:50, 50:100, 0] = 255
#image[0:50, 50:100, 2] = 255
plt.imshow(image)
```

```python
print (image[50:100, 100:150,0])
```

recortar imagen

```python
lena_face = image[100:200, 180:260, : ]
plt.imshow(lena_face)
print (lena_face.shape)
```

```python
image = Image.open("/content/lena.jpg")
image = np.array(image)
image[100:200, 300:320, : ] = lena_face
plt.imshow(image)
```
